```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load data
data = pd.read_csv('best-fit.csv')
data.head()
```

```python
# Plotting the real data points 
data.plot(x='X', y='Y', style='o')
plt.title('x vs y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
# performing the linear regression using polyfit function
m, b = np.polyfit(data.X, data.Y, 1)

# Plotting the regression line
plt.plot(data.X, data.Y, 'o')
plt.plot(data.X, m*data.X + b)

plt.title('x vs y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```
